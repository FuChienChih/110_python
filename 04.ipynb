信用卡號碼驗證

信用卡號碼遵循特定形式。信用卡號碼必須含有13~16位數。起始位數必須是： 
 4代表Visa cards 
 5代表Master cards 
 37代表American Express cards 
 6代表Discover cards 
1954年，IBM的Hans Luhn提出一個驗證信用卡號碼的演算法。此演算法用在判斷輸入的卡號是否正確，或在信用卡是否正確被掃描上很有用。信用卡號遵循此來產生有效驗證，這種作法稱為Luhn check或Mod 10 check，說明如下(為了方便解釋，這裡以卡號4388576018402626為例)： 


1. 由左到右，每隔一位數便取其兩倍數字。如果該位數的兩倍數字為兩位數，便將該兩位數相加，以取單一位數數字。
     請參考附件圖 
2. 接著將步驟1所有的單位數數字相加。
4+4+8+2+3+1+7+8=37 
3. 再由右到左，將出現於卡號奇數位置的所有位數相加。
6+6+0+8+0+7+8+3=38 
4. 將步驟2與步驟3的結果相加。
37+38=75

5. 如果步驟4的結果能夠被10整除，則卡號便是有效的；反之，則是無效的。
比方說，卡號4388576018402626是無效的，但卡號4388576018410707是有效的。 
請撰寫一程式，提示使用者輸入信用卡卡號為整數， 接著顯示該號碼是否為有效卡號。

```python
import sys
card_number = input('輸入信用卡卡號為整數')
card_number_list = list(map(int,card_number))
# 防呆:
if len(card_number_list)>16 or len(card_number_list)<13:
    sys.exit('信用卡號碼必須含有13~16位數')
new_number = list() # 新數
def new_number_funtion(card_number_list):
    for i in range(len(card_number_list)):
        if i%2 == 0: # 要取兩倍的數字
            if card_number_list[i]*2 >=10 :#若此號碼的兩倍>=10
                new_number.append(card_number_list[i]*2-10+1)
            else :
                new_number.append(card_number_list[i]*2)
        else : # 維持原數字的數字
            new_number.append(card_number_list[i])
new_number_funtion(card_number_list)

# 判斷是否為有效卡號
if sum(new_number)%10 == 0:
    print('有效卡號')
else:
    print('無效卡號')
```
