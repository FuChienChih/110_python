```python
def per_line(marker , num):
    for i in range(10 - num//2):
        print(' ',end='')
    for i in range(num):
        print(marker,end='')
    print('')

def special():
    print('      ###   ###')
```

```python
num_list = [1,3,1,1] # 每行符號的數量
marker = '+'
for num in num_list:
    per_line(marker , num)
num_list = [1,3,5,7,9,7,9,11,13,11,13,15,17,15,17,19,21]
marker = '*'
for num in num_list:
    per_line(marker , num)
num_list = [9,9]
marker = '#'
for num in num_list:
    per_line(marker , num)
special()
special()
for num in num_list:
    per_line(marker , num)
```
