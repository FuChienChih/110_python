試著利用迴圈來輸出空心矩形吧

```python
# 畫圖函示
def rectangle(x,y):
    for i in range(y):
        if i == 0 or i == y-1:
            for j in range(x):
                print('* ',end='')
            print('')
        else :
            for j in range(x):
                if j == 0 or j == x-1:
                    print('* ',end='')
                else :
                    print('  ',end='')
            print('')

# 主程式
x,y = input().split(' ')
rectangle(int(x),int(y))
```
